## Combine Age Data into one spreadsheet

```python
age2534 = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Estimated age 25 to 34.xlsx',sheet_name='Age 25 to 34')
age3544 = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Estimated age 35-44.xlsx',sheet_name='Age 35 to 44')
age4554 = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Estimated age 45 to 54.xlsx',sheet_name='Age 45 to 54')
age5564 = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Estimated age 55 to 64.xlsx',sheet_name='Age 55 to 64')
age65 = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Estimated age 65 or older.xlsx', sheet_name='People age 65 or older')
agemed = pd.read_excel('Data given by Lisa/Census Tract Level/Age/Median Age.xlsx', sheet_name = 'Median Age')
```

```python
age2534 = age2534.drop(index=range(74134,74145))
age3544 = age3544.drop(index=range(74134,74145))
age4554 = age4554.drop(index=range(74134,74145))
age5564 = age5564.drop(index=range(74134,74145))
agemed = agemed.drop(index=range(74134,74145))
```

```python
def merger(*args):
    new = args[0]
    for i in range(len(args)-1):
        new = pd.merge(new,args[i+1],how = 'inner')
    return new
age = merger(age2534,age3544,age4554,age5564,agemed)
```

## For each index(Zip_1), we want the average of all the data belonging to this zipcode

```python
df = d3_temp.merge(age, how = 'left', left_on = 'tract', right_on = 'Census Tract').dropna().set_index('Zip_1').sort_index()
```

```python
df.groupby(df.index).agg('mean').drop(['tract','FIPS Code', 'Formatted FIPS',],axis=1)
```
