### 4.2 Model on data_comp

- This model bases on a dataset where every patient has prescription, but not necessarily has paid claims.
- If we research whether they have pd_claims after prescription, it can check the factors that affect the patient to reverse/get rejected

```python
[col for col in data_comp.columns if data_comp[col].dtypes in ['float64','int64','int32']]
```

```python
num_cols = [col for col in data_comp.columns if data_comp[col].dtypes in ['float64','int64','int32']]

DV_cols = [
    'total_pd_claims',
    'total_pd_saxenda_claims',
    'total_pd_contrave_claims',
    'total_pd_qsymia_claims',
    'total_pd_belviq_claims',
    'total_pd_generic_claims',
    'pd_claims_yn'
]

for col in DV_cols:
    num_cols.remove(col)

num_cols.remove('patient_id')
num_cols.remove('patient_birth_year')
num_cols.remove('joined_prescriber_id')
num_cols.remove('zip')
num_cols.remove('joined_plantrak_id')
num_cols.remove('days_between_first_pdprescr_latest_pdprescr')

num_cols
# feature_cols = 
```

```python
cat_cols = [col for col in data_comp.columns if not data_comp[col].dtypes in ['float64','int64','int32']]
cat_cols
```

```python
y = data_comp.pd_claims_yn
X = data_comp[num_cols+cat_cols]
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X,y,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ])
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error

# Bundle preprocessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                             ])

# Preprocessing of training data, fit model 
my_pipeline.fit(X_train, y_train)

# Preprocessing of validation data, get predictions
preds = my_pipeline.predict(X_valid)

# Evaluate the model
score = mean_absolute_error(y_valid, preds)
print('MAE:', score)
```
