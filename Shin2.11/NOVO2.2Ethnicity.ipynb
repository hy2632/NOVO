## Combine Ethnicity Data into one spreadsheet

```python
asian = pd.read_excel('/Users/shinler/Documents/Columbia MSE/OpsCon/NN/Socio-Economic Data/Ethnicity/Estimated Asian Percentage 2013 to 2017.xlsx',sheet_name="Estimated Asian pop")
black = pd.read_excel('/Users/shinler/Documents/Columbia MSE/OpsCon/NN/Socio-Economic Data/Ethnicity/Estimated Black Percentage 2013 to 2017.xlsx',sheet_name='Black percentage pop')
hispanic = pd.read_excel('/Users/shinler/Documents/Columbia MSE/OpsCon/NN/Socio-Economic Data/Ethnicity/Estimated Hispanic Percentage 2013 to 2017.xlsx',sheet_name='Estimated Hispanic pop')
white = pd.read_excel('/Users/shinler/Documents/Columbia MSE/OpsCon/NN/Socio-Economic Data/Ethnicity/Estimated White Percentage 2013 to 2017.xlsx',sheet_name='Estimated White pop')
```

```python
def merger(*args):
    new = args[0]
    for i in range(len(args)-1):
        new = pd.merge(new,args[i+1],how = 'inner')
    return new
ethnicity = merger(asian,black,hispanic,white)
```
