```python
df_eth = pd.read_excel('Ethnicity.xlsx')
df_eth
```

```python
df_age = pd.read_csv('Age_Output.csv')
df_age
```

```python
df_hhs = pd.read_excel('HHSpending.xlsx')
df_hhs
```

```python
def merger(*args):
    new = args[0]
    for i in range(len(args)-1):
        new = pd.merge(new,args[i+1],how = 'inner')
    return new
df_total = merger(df_age,df_edu,df_eth,df_hhs)
df_total
```

```python
df_total = df_total.drop_duplicates().set_index('zip')
```

```python
dict_zip_city[['ZipCode','City','State']].merge(df_total,left_on = 'ZipCode', right_on = df_total.index, how = 'right')
```

```python
dict_zip_city[['ZipCode','City','State']].merge(df_total,left_on = 'ZipCode', right_on = df_total.index, how = 'right')\
    .reset_index().set_index('ZipCode').to_csv('AgeEduEthHHS.csv')
```
