```python
df_2['sum'] = df_2['Asian']+df_2['Black']+df_2['Hispanic']+df_2['White']
```

```python
df_2['Asian']=df_2['Asian']/df_2['sum']*100
df_2['Black']=df_2['Black']/df_2['sum']*100
df_2['Hispanic']=df_2['Hispanic']/df_2['sum']*100
df_2['White']=df_2['White']/df_2['sum']*100
```

```python
df_2
```

```python
df_1.merge(df_2[['zip','Asian','Black','Hispanic','White','Predominant']], left_on = 'ZipCode', right_on = 'zip',how='left').drop('zip',axis=1).set_index('ZipCode')
```

```python
df_3 = df_1.merge(df_2[['zip','Asian','Black','Hispanic','White','Predominant']], left_on = 'ZipCode', right_on = 'zip',how='left').drop('zip',axis=1)
```

## 2.16 Add Zip Code population Data

```python
df_zipcode_population = pd.read_csv('2.16 ZipCode Population/zip-codes-database-FREE.csv')
```

```python
df_zipcode_population
```

```python
df_3 = df_3.merge(df_zipcode_population[['ZipCode','Population']], on = 'ZipCode', how = 'left').dropna()
df_3
```

```python
df_3.ZipCode = df_3.ZipCode.astype('str')
df_3.ZipCode  = df_3.ZipCode.apply(lambda x:('00'+x)[-5:] if len(x) <= 4 else x)
```
