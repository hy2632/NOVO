## Combine Ethnicity Data into one spreadsheet

```python
ethnicity = pd.read_excel('/Users/shinler/Documents/Columbia MSE/OpsCon/NN/Socio-Economic Data/Ethnicity/Combined Ethnicity.xlsx',sheet_name="Estimated Asian pop")
```

## We can use Census Tract as Index

```python
ethnicity = ethnicity.set_index('Census Tract')
```

```python
ethnicity = ethnicity.dropna()
```

```python
ethnicity
```

## Change data type(object to int)

```python
ethnicity.index = ethnicity.index.astype('int64')
```

```python
ethnicity.dtypes
```

```python
ethnicity = ethnicity.sort_index()
```

```python
ethnicity
```

## Just use the newest dict

```python
d2018['zip'] = d2018['zip'].astype('str')
```

```python
d2018['zip']  = d2018['zip'].apply(lambda x:('00'+x)[-5:] if len(x) <= 4 else x)
```

```python
d2018_temp = d2018.reset_index()
```

```python
d2018_temp
```

```python
d2018_temp.merge(ethnicity, how = 'left', left_on = 'tract', right_on = 'Census Tract')
```

## Looks like there is some missing data (dict > ethnicity Data)

```python
d2018_temp.merge(ethnicity, how = 'left', left_on = 'tract', right_on = 'Census Tract').dropna()
```

## Reset index to zipcode

```python
d2018_temp.merge(ethnicity, how = 'left', left_on = 'tract', right_on = 'Census Tract').dropna().set_index('zip').sort_index()
```

## For each index(Zip_1), we want the average of all the data belonging to this zipcode
 - we do not want unrelated columns to be averaged so just drop them.

```python
df = d2018_temp.merge(ethnicity, how = 'left', left_on = 'tract', right_on = 'Census Tract').dropna().set_index('zip').sort_index()
```

```python
df = df.groupby(df.index).agg('mean').drop(['tract',],axis=1)
```
