```python
import pandas as pd
```

```python
df_1 = pd.read_csv('AgeEduEthHHS.csv')
df_2 = pd.read_excel('Ethnicity.xlsx')
```

```python
df_1 = df_1.drop('index',axis=1)
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_1.merge(df_2[['zip','Predominant']], left_on = 'ZipCode', right_on = 'zip',how='left').drop('zip',axis=1).set_index('ZipCode')
```

```python
df_final = df_1.merge(df_2[['zip','Predominant']], left_on = 'ZipCode', right_on = 'zip',how='left').drop('zip',axis=1)
```

```python
df_final.ZipCode = df_final.ZipCode.astype('str')
df_final.ZipCode  = df_final.ZipCode.apply(lambda x:('00'+x)[-5:] if len(x) <= 4 else x)
```

```python
df_final.set_index('ZipCode')
```

```python
df_final.set_index('ZipCode').to_csv('FinalSEData.csv')
```
